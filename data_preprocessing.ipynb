```python
import pandas as pd
```

```python
matchlist_df = pd.read_csv('data\matchlist.csv')
```

```python
matchlist_df.head()
```

```python
#Converting match ID into numeral.

matchlist_df['Scorecard'] = matchlist_df['Scorecard'].str.replace('T20I # ', '', regex=True)
```

```python
matchlist_df.head()
```

```python
matchdetails_df = pd.read_csv('data\matchdetails.csv')
```

```python
matchdetails_df.head()
```

```python
#Converting match ID into numeral.

matchdetails_df['MatchId'] = matchdetails_df['MatchId'].str.replace('T20I no. ', '', regex=True)
```

```python
matchdetails_df.head()
```

```python
#Splitting toss result into two columns.
matchdetails_df[['Toss Winner', 'Decision']] = matchdetails_df['TossResult'].str.split(', ', expand=True)
```

```python
matchdetails_df.head()
```

```python
matchdetails_df.drop('TossResult',axis=1,inplace=True)
```

```python
matchdetails_df.head()
```

```python
MatchDetails = pd.merge(matchlist_df, matchdetails_df, left_on='Scorecard',right_on='MatchId',how='inner')
```

```python
MatchDetails.head()
```

```python
#Dropping duplicate column
MatchDetails=MatchDetails.drop('Scorecard',axis=1)
```

```python
MatchDetails.head()
```

```python
MatchDetails.rename(columns={'MatchId':'Match ID', 'Name':'POTM'}, inplace=True)
```

```python
MatchDetails.head()
```

```python
MatchDetails.isna().sum()
```

```python
MatchDetails.describe()
```

```python
playerdetails_df = pd.read_csv('data/playerdetails.csv')
```

```python
playerdetails_df.head()
```

```python
#Let's make amendsments in caps.
playerdetails_df['Cap'] = playerdetails_df['Cap'].str.replace('(vc)', 'VC')
playerdetails_df['Cap'] = playerdetails_df['Cap'].str.replace('(c)', 'C')
playerdetails_df['Cap'] = playerdetails_df['Cap'].str.replace('†', '')
```

```python
# Removing Squad from country
playerdetails_df['Country'] = playerdetails_df['Country'].str.replace(' Squad','')
```
